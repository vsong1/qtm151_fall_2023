(b) Counting True/False statements in a list

- Create
list_age = [58, 23, 63, 18, 76, 35, 48, 68]
- Create an empty list "list_eligible_retire"
- Use a for-loop without numbering
    - Determine whether the age is $\ge 65$
    - Store the True/False result in "list_eligible_retire"
- Use "np.sum()" to count how many workers are eligible for retirement

Hint: Use the "append" function.

```python
list_age = [58, 23, 63, 18, 76, 35, 48, 68] 
list_eligible_retire = [] #empty list

for age in list_age:
    if age >= 65: 
        list_eligible_retire.append(True)
    else: 
        list_eligible_retire.append(False)

np.sum(list_eligible_retire)
#Two workers are eligible to retire.
```

- Create three lists

list_lower = [93, 90, 87, 83, 80, 77, 73, 70, 67, 63, 60, 0]

list_upper = [100, 92, 89, 86, 82, 79, 76, 72, 69, 66, 62, 59]

list_letters = ["A","A-","B+","B","B-","C+","C","C-","D+","D","D-","F"]

- Create a for loop that prints the message:

"If a students gets a grade between .... and ..., their letter grade is ..."

```python
list_lower = [93, 90, 87, 83, 80, 77, 73, 70, 67, 63, 60, 0]
list_upper = [100, 92, 89, 86, 82, 79, 76, 72, 69, 66, 62, 59]
list_letters = ["A","A-","B+","B","B-","C+","C","C-","D+","D","D-","F"]

for i in range(len(list_letters)):
    print("If a student gets a grade between "+str(list_lower[i])+" and "+str(list_upper[i])+", their letter grade is "+str(list_letters[i])+".")
```

(e) Classification with double loops

<font size ="3">

- Suppose that 7 students get the following grades:

list_studentgrades = [84, 62, 92, 74, 97, 80, 79]

- Print the following message:

"The student got a numeric grade of .... and their letter grade was ..."


```python
list_studentgrades = [84, 62, 92, 74, 97, 80, 79] #list of grades

for grade in list_studentgrades: #iterate through each student's grade
    index = 0 #index for searching list_lower
    for letter in list_letters: #iterates through letter grades from A to F
        #checks range for each letter grade
        if grade >= list_lower[index] and grade <= list_upper[index]:
            #if grade is within this letter's range, print statement and break the inner 'for' loop
            print("The student got a numeric grade of "+str(grade)+" and their letter grade was "+letter+".")
            break
        else:
            #if grade is not within this letter's range, increment index by 1 and continue to next iteration of loop
            index = index + 1
```
