(a) Compute function for each individual (map)

<font size = "3">

- Import the datset "data/bmi.csv"
    - **Note that the column `Height` is in $cm$, and `Weight` is in $kg$**
- Write a lambda function for $ BMI = \frac{weight (kg)}{height (m)^2} $
- Use ```list(map(...))``` to calculate $BMI$ for each row of the dataset
- Round the outcome into 2 decimal points
- Store the calculation in a new column in the dataframe called `BMI`
- Display the dataframe


````python
bmi_df = pd.read_csv("data/bmi.csv") #import dataframe

bmi_lambda = lambda w, h: w/((h/100)**2) #lambda function; h/100 is height in meters

bmi = list(map(bmi_lambda, bmi_df["Weight"], bmi_df["Height"]))

bmi = [round(i, 2) for i in bmi] #iterate through and round each entry in list

bmi_df["BMI"] = bmi #create new column in dataframe and store list as column

display(bmi_df) #display dataframe as table
````

- Print the original `countryname` and the newly created column, like: <br>
"['Afghanistan', 'Africa Eastern and Southern', 'Africa Western and Central', 'Albania', 'Algeria', ......] "

````python
wdi_2020 = pd.read_csv("data/wdi_2020.csv")

fn_upper = lambda x: x.upper()

wdi_2020["countryname_upper"] = list(map(fn_upper, wdi_2020["countryname"]))

print(list(wdi_2020["countryname"]))
print(list(wdi_2020["countryname_upper"]))
````

(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
wdi_urban = pd.read_csv("data/wdi_urban.csv")

years = list(wdi_urban["year"].drop_duplicates())

for yr in years:
    plt.hist(wdi_urban.query('year == @yr')["prop_urbanpopulation"], alpha = 0.5)
plt.legend(labels = years, loc = "lower right")
plt.xlabel("Proportion of Total Population Living in Urban Areas (%)")
plt.ylabel("Number of Countries")
plt.title("Proportion of Urban Populations in Various Countries")

# Looking at the histogram, it appears that the proportion of countries living in urban areas has increased from 1980 to 2020.
````
